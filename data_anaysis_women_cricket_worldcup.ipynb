Batter TYpes

```python
def batter_type(row):
    if row['Average'] > 40 and row['Strike_rate'] > 90:
        return 'Elite'
    elif row['Average'] > 40:
        return 'Anchor'
    elif row['Strike_rate'] > 90:
        return 'Power Hitter'
    else:
        return 'Inconsistent'

bat['Batter_Type'] = bat.apply(batter_type, axis=1,)
bat

```

NOW SOME ANLYSIS FOE WINNING TEAM INDIA


How many match India played?

```python
india_matches = matches[
    (matches["team1"] == "India") | (matches["team2"] == "India")
]

total_matches = india_matches.shape[0]
print("Total Matches Played by India:", total_matches)
```

how many India won?

```python
india_wins = india_matches[
    india_matches["winning_team"] == "India"
].shape[0]

print("Matches Won by India:", india_wins)

```

win percentage

```python
win_percentage = (india_wins / total_matches) * 100
loss_percentage = 100 - win_percentage

print("Win Percentage:", round(win_percentage,2), "%")
```

pie chart over win and loss of team India

```python
labels = ["Wins", "Losses"]
sizes = [win_percentage, loss_percentage]

plt.figure()
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title("India Overall Win Percentage")

plt.show()
```

Bar chart over win and loss of Indian Team

```python
data = {
    "Result": ["Wins", "Losses"],
    "Percentage": [win_percentage, loss_percentage]
}

df = pd.DataFrame(data)

plt.figure()
plt.bar(df["Result"], df["Percentage"])
plt.title("India Win vs Loss Percentage")
plt.ylabel("Percentage")

plt.show()
```
